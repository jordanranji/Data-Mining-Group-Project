```python
#Accidentally added the play by play data to a new dataframe instead of appending, combining here

import pandas as pd
```

```python
pbp = pd.read_csv("../data/PlayByPlay.csv")
df = pd.read_csv("../data/PBPData.csv")



correctish = df.loc[df.away_lineup_id.isna()]

df = df.loc[~df.away_lineup_id.isna()][pbp.columns]

correctish = correctish.rename(columns={correctish.columns[i]: correctish.columns[i + 2] for i in range(len(correctish.columns) - 2)})
correct = correctish.iloc[:, :-2]
#Correct is correct now
```

```python
pbps = pd.concat([pbp, df, correct])
```
